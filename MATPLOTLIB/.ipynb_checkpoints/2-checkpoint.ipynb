# scatter graph

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# scatter graph is shows realationship differnce two variables
# it shows correlation +,-,None
#for example: mark vs study hours
```

```python
x1 = [1,2,3,4,5,6,6,7,8,9]
y1 = [35,45,51,57,62,68,71,88,93,97]

x2 = [1,1,2,3,4,5,5,6,7,8,9]
y2 = [25,37,41,49,52,68,71,82,87,91,99]

plt.title("Test Series")
plt.xlabel("Study hour")
plt.ylabel("marks")
plt.scatter(x1,y1,
            label="Class A",
            alpha=0.5,
            color="red"
           )
plt.scatter(x2,y2,
            label="Class B",
            alpha=0.5,
            color="blue"
           )
plt.legend()
plt.show()
```

# Histogram

```python
# Distributed numerical data
# it group value into ranges(bins)

#Is the data normally distributed?
#Where are most values concentrated?
#Are there any outliers?
#Is the data skewed left or right?
#How spread out is the data?


#| Parameter | Meaning                                         |

# `loc`     | Center of the distribution (mean)               
# `scale`   | Spread of the distribution (standard deviation) 
# `size`    | Number of random values to generate             
```

```python
marks = [45, 50, 52, 54, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 82, 85, 88, 90]

data = np.random.normal(loc=80,size=10,scale=5)
plt.hist(marks,color="skyblue",edgecolor="black")
plt.show()
```

# Subplots

```python
# figure = show entire canvas
# ax = single plot(sub plot)
```

```python
figure,axis = plt.subplots(2,2) #give as set

x1 = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]
axis[0,0].plot(x1,y1,color="red")
axis[0,0].set_title("plot1")


x2 = [10, 20, 30, 40]
y2 = [5, 15, 16, 20]
axis[0,1].plot(x2,y2,color="blue")
axis[0,1].set_title("plot2")


x3 = [0, 5, 10, 15]
y3 = [100, 80, 60, 40]
axis[1,0].scatter(x3,y3,color="skyblue")
axis[1,0].set_title("plot3")


x4 = [2, 4, 6, 8]
y4 = [3, 7, 4, 9]
axis[1,1].bar(x4,y4,color="green")
axis[1,1].set_title("plot4")

plt.tight_layout()


plt.show()
```
