# plot

```python

x = np.array([2001,2002,2003,2004,2005])

list_attr = dict(
    marker=".",
    markersize=17,
    markerfacecolor="red",
    linestyle = "dotted",
    linewidth=4
)
y1 = np.array([67,30,35,40,29])
y2 = np.array([40,20,26,17,54])
y3 = np.array([34,45,56,78,11])


plt.title("Birth by year",
         fontsize=20,
         #family="ans-serif",
         fontweight="bold",
         color="darkblue"
)

plt.xlabel(
    "year",
    fontsize=20,
    fontweight="bold",
    color="#171ee3"
)

plt.ylabel(
    "Birth",
    fontsize=20,
    fontweight="bold",
    color="#0ac723"
)

plt.xticks(x)
plt.tick_params(axis="both", colors="#0698d6")

plt.plot(x,y1,color="#cf1127", **list_attr) #** open the dict
plt.plot(x,y2,color="#132fba", **list_attr) #** open the dict
plt.plot(x,y3,color="#1da396", **list_attr) #** open the dict
```
