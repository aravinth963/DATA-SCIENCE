```python
#30. Write a program to count the number of occurrences of each element in a list using a dictionary.

numbers = [1,1,2,2,3,3,3,5,6,9]

dict1 = {}

for i in numbers:
    if i in dict1:
        dict1[i] +=1
    else:
        dict1[i] = 1
print(dict1)
```
