```python
import pandas as pd
```

```python
fruits = ["Apple", "Banana", "Orange", "Mango"]
series = pd.Series(fruits)
print(series.tolist())
```

```python
Values = [10, 20, 30, 40]
Index = ["A", "B", "C", "D"]
series = pd.Series(Values, index = Index)

print(series.loc["C"])

print(series.head(2))
print(series.tail(2))
print(series.sample(3))
```

```python
s = pd.Series([5, 10, 15, 20, 25])

print(s.iloc[0])
print(s.iloc[-1])
print(s.sum())
```

```python
s = pd.Series([12, 7, 25, 18, 30])

print(s[s > 15].tolist())
print(s.shape)
```
