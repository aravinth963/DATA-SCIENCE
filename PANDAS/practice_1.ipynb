```python
import pandas as pd

data = {
    "Name":["Arun","Bala","Charan","Divya","Esha"],
    "Age":[21,22,20,23,21],
    "Department":["IT","HR","IT","Sales","HR"],
    "Salary":[45000,35000,50000,42000,37000]
}

df = pd.DataFrame(data)
```

```python
#Display the last 2 rows.
df.tail(2)
```

```python
#Print only the Name column.
df["Name"]
```

```python
#Print Name and Salary.

df[["Name","Salary"]]
```

```python
#Find the shape of the dataframe.

df.shape
```

```python
#Display column names.

df.columns
```

```python
#Find the data type of every column.

df.dtypes
```

```python
#Print all employees whose salary is greater than 40000.

df[df["Salary"] > 40000]
```

```python
#Print employees from the IT department.

df[df["Department"] == "IT"]
```

```python
#Find the average salary.

df["Salary"].mean() #41800.00
int(df["Salary"].mean())
```

# Level 2

```python
#Employees whose age is greater than 21.

df[df["Age"] > 21]
```

```python
#Employees whose department is HR.

df[df["Department"] == "HR"]
```

```python
#Employees with salary between 35000 and 45000.

df[(df["Salary"] > 35000) & (df["Salary"] < 45000)]
```

```python
#Employees whose department is IT and salary > 45000.

df[(df["Department"]=="IT") & (df["Salary"] > 45000)]
```

```python
#Employees whose department is HR or Sales.

df[(df["Department"] == "HR") | (df["Department"] == "Sales")]
```

```python
#Employees whose name starts with "A".

df[df["Name"].str.startswith("A")]
```

```python
#Employees whose name ends with "a".

df[df["Name"].str.startswith("a")]
```

```python
#Employees whose age is 21 or 23.

df[(df["Age"] == 21) | (df["Age"] == 23)]
```

```python
#Find employees whose salary is not equal to 42000.

df[df["Salary"] != 42000]
```

```python
#Sort employees by salary in ascending order.

df.sort_values("Salary",ascending=False)
```
