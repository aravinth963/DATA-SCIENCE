# Level 1 (Basics)

# Level 2 (Filtering)

```python
#Employees whose age is greater than 21.

df[df["Age"] > 21]
```

```python
#Employees whose department is HR.

df[df["Department"] == "HR"]
```

```python
#Employees whose name ends with "a".

df[df["Name"].str.startswith("a")]
```

# Level 6 (Missing Values)

```python
data = {
    "Name":["A","B","C","D","E"],
    "Age":[20,None,21,22,None],
    "Salary":[30000,35000,None,45000,50000]
}

df = pd.DataFrame(data)
```

```python
#Find missing values.

df.isna()
df.isna().any().any()
```

```python
#Count missing values.

df.isna().sum()
```

```python
#Drop rows with missing values.

df.dropna()
```

```python
#Fill missing age with average age.
avg_age = df["Age"].mean()
df["Age"]=df["Age"].fillna(avg_age)
df
```

```python
#Fill missing salary with median salary.
avg_salary = df["Salary"].median()
df["Salary"]=df["Salary"].fillna(avg_salary)
df
```

```python
#Replace all missing values with 0.

df.fillna(0)
```

```python
#Check if dataframe has any missing value.

df.isna().values.any()
```

```python
#Find percentage of missing values in each column.

(df.isna().sum()/len(df))*100
df.isna().mean() * 100
```

```python
#Drop columns having missing values.

df.dropna(axis=1) # 0 means rows,1 means columns
```

```python
#Drop columns having missing values.

df.dropna(axis=0) # 0 means rows,1 means columns
```

```python
df
```

```python
#Fill missing values using forward fill.

df.ffill().bfill()
```
