```python
import pandas as pd

data = [101,102,103]

series = pd.Series(data, index=["a","b","c"])

print(series)
```

```python
data = [101,102,103,200,204]

series = pd.Series(data, index=["a","b","c","d","e"])

print(series[series >= 200])
```

```python
#--------------------
#Dictionary
#--------------------
```

```python
import pandas as pd

calories = {"Day1" : 1750,"Day2" : 2000,"Day3" : 1700}

series = pd.Series(calories)
print(series.loc["Day1"])
print(series.iloc[0])

series.loc["Day3"] += 500

print(series.loc["Day3"])

print(series[series < 2000])
print(series[series >= 2000])
```

```python
import pandas as pd

name = {1:"Bulbasaur",
        2:"Ivysaur",
        3:"Venusaur",
        4:"Charmander",
        5:"Charmelon",
        6:"Charizard"}

series = pd.Series(name)

print(series)
```
